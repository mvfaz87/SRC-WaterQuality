```python
import pandas as pd
```

```python
df = pd.read_excel('data/WaterQualityTable.xlsx')
```

```python
df
```

```python
df.drop(columns = 'Unnamed: 0', inplace=True)
df['Sample_Date'] = pd.to_datetime(df['Sample_Date']).dt.date
```

```python
df.set_index(['Site','Sample_Date'], inplace = True)
```

```python
df1 = df.groupby(['Site','Sample_Date']).count()
```

```python
df2 = df.groupby('Site').count()
```

```python
df1.to_excel('Data/site_by_date_count.xlsx')
```

```python
df2.to_excel('Data/site_count.xlsx')
```
